```python
#function to cnvert ip address to  num
def ip_to_int(ip_ser):
    ips = ip_ser.str.split('.', expand=True).astype(np.int16).values
    mults = np.tile(np.array([24, 16, 8, 0]), len(ip_ser)).reshape(ips.shape)
    return np.sum(np.left_shift(ips, mults), axis=1)
```

```python
trainx['_ip'] = ip_to_int(trainx.IP_Address)
```

```python
function to convert ip to num
import socket, struct

def ip2long(ip):
    """
    Convert an IP string to long
    """
    packedIP = socket.inet_aton(ip)
    return struct.unpack("!L", packedIP)[0]
```

```python
ip_boundaries["lower_bound_ip_address"] = ip_boundaries["lower_bound_ip_address"].apply(ip2long)
ip_boundaries["upper_bound_ip_address"] = ip_boundaries["upper_bound_ip_address"].apply(ip2long)
```
